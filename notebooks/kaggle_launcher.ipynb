```python
# Cell 4: Override Config Paths for Kaggle
import yaml

CONFIG_PATH  = 'configs/config.yaml'
DATASET_ROOT = '/kaggle/input/satellite-images-of-water-bodies/Water Bodies Dataset'
WORKING_DIR  = '/kaggle/working/water-body-segmentation-in-satellite-images'

with open(CONFIG_PATH, 'r') as f:
    cfg = yaml.safe_load(f)

cfg['paths']['image_dir']        = f'{DATASET_ROOT}/Images'
cfg['paths']['mask_dir']         = f'{DATASET_ROOT}/Masks'
cfg['paths']['valid_files']      = f'{WORKING_DIR}/data/valid_files.txt'
cfg['paths']['checkpoint_dir']   = '/kaggle/working/outputs/checkpoints'
cfg['debug']['enabled']          = False
cfg['checkpoint']['resume']      = False
cfg['checkpoint']['resume_path'] = '/kaggle/working/outputs/checkpoints/last_checkpoint.pth'

with open(CONFIG_PATH, 'w') as f:
    yaml.dump(cfg, f, default_flow_style=False)

print('Config updated for Kaggle:')
print(f"  image_dir      : {cfg['paths']['image_dir']}")
print(f"  mask_dir       : {cfg['paths']['mask_dir']}")
print(f"  valid_files    : {cfg['paths']['valid_files']}")
print(f"  checkpoint_dir : {cfg['paths']['checkpoint_dir']}")
print(f"  debug          : {cfg['debug']['enabled']}")
print(f"  model          : {cfg['model']['name']}")
print(f"  epochs         : {cfg['train']['epochs']}")
print(f"  run_name       : {cfg['wandb']['run_name']}")
```
